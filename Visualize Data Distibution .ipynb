```python
irisdata = pd.read_csv('C:\\Users\\User\\Documents\\basicpython\\python_dataviz\\iris.data',header=None)
irisdata
```

เอาแถวที่ 0 ถึง 50 ตัดคอมลัมที่ 1 ออกไป

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60)
```

plot data 3 หลัก: ข้อมูลหลักที่1= X ,หลักที่สอง= Y ,หลักที่สาม= ขนาด

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2])
```

การเปลี่ยนสีข้อมูลด้วย c=' '

```python
plt.scatter(setosa_np[:,0],setosa_np[ : ,1],s=10*setosa_np[:,2], c = 'r')
```

```python
print(f'max = {max(setosa_np[:,3])}, min = {min(setosa_np[:,3])}')
```

สร้าง list เพื่อเปลี่ยนตัวเลขเป็นตัวอักษรตามสี

```python
colors = []
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif i < 0.3:
        colors.append('g')
    elif i < 0.45:
        colors.append('y')
    else:
        colors.append('r')
colors                   
```

```python
plt.scatter(setosa_np[:,0],setosa_np[ : ,1],s=10*setosa_np[:,2], c = colors)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[ : ,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap='YlOrBr')
```

```python
plt.scatter(setosa_np[:,0],setosa_np[ : ,1],s=30*setosa_np[:,2], c = np.array([[1,0.5,0.8]]))
```

```python
plt.scatter(setosa_np[:,0],setosa_np[ : ,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap='YlOrBr',marker ='s')
```
