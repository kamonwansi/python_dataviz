```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel('C:\\Users\\User\\Documents\\basicpython\\python_dataviz\\Average Debt.xls',header=3) #header=3คือไม่สนใจ 3 แถวแรก
df
```

```python
greaterbkk = df.iloc[4:8,:]
greaterbkk
```

## Barchart กราฟแท่ง

### สร้างกราฟแท่งเปรียบเทียบรายได้ในปี 2541 ของจังหวัดกรุงเทพและปริมณฑล

```python
plt.bar([1,2,3,4],greaterbkk.iloc[:,2])
plt.title('Barchart comparing average monthly income of provinces in Greater Bangkok in 2541')
```

ใส่ชื่อให้กับแต่ละแกน

```python
plt.bar([1,2,3,4],greaterbkk.iloc[:,2])
plt.title('Barchart comparing average monthly income of provinces in Greater Bangkok in 2541')
plt.xlabel('Provinces')
plt.ylabel('Income Bath/Month/Household')
```

## Add tick label

```python
plt.bar([1,2,3,4],greaterbkk.iloc[:,2],tick_label = greaterbkk.iloc[:,-1])
plt.title('Barchart comparing average monthly income of provinces in Greater Bangkok in 2541')
plt.xlabel('Provinces')
plt.ylabel('Income Bath/Month/Household')
```

```python
plt.bar([1,2,3,4],greaterbkk.iloc[:,2],tick_label = greaterbkk.iloc[:,-1],color = 'rygb',edgecolor = 'black') ##color คือ เติมสีกราฟแท่ง, edgecolor คือ สีของกรอบกราฟแท่ง
plt.title('Barchart comparing average monthly income of provinces in Greater Bangkok in 2541')
plt.xlabel('Provinces')
plt.ylabel('Income Bath/Month/Household')
```

## compare different years

##### width = กำหนดขนาดกราฟ , +/- = กำหนดการเลื่อนตำแหน่งของกราฟ

```python
plt.bar([1-((0.35)/2),2-((0.35)/2),3-((0.35)/2),4-((0.35)/2)],greaterbkk.iloc[:,-2],width = 0.35,tick_label = greaterbkk.iloc[:,-1],color = 'black',label = '2558')
plt.bar([1+((0.35)/2),2+((0.35)/2),3+((0.35)/2),4+((0.35)/2)],greaterbkk.iloc[:,2],width = 0.35,tick_label = greaterbkk.iloc[:,-1],color = 'yellow',label = '2541')
plt.title('Barchart comparing average monthly income of provinces in Greater Bangkok in 2541,2558')
plt.xlabel('Provinces')
plt.legend()
plt.ylabel('Income Bath/Month/Household')
```

# Stacked Barchart

```python
saledata = pd.DataFrame([{'North':20,'East':10,'West':90,'South':120},
                                    {'North':220,'East':20,'West':50,'South':90}])
```

```python
saledata
```

```python
plt.bar([1,2],saledata.iloc[:,0],tick_label=['Company A' , 'Company B'])
plt.bar([1,2],saledata.iloc[:,1],bottom = saledata.iloc[:,0],tick_label=['Company A' ,'Company B' ])
plt.bar([1,2],saledata.iloc[:,2],bottom = saledata.iloc[:,0],tick_label=['Company A' ,'Company B' ])
```

```python
plt.bar([1,2],saledata.iloc[:,0],tick_label=['Company A' , 'Company B'])
plt.bar([1,2],saledata.iloc[:,1],bottom = saledata.iloc[:,0],tick_label=['Company A' ,'Company B' ])
plt.bar([1,2],saledata.iloc[:,2],bottom = saledata.iloc[:,0]+saledata.iloc[:,1],tick_label=['Company A' ,'Company B' ])
plt.bar([1,2],saledata.iloc[:,3],bottom = saledata.iloc[:,0]+saledata.iloc[:,1]+saledata.iloc[:,2],tick_label=['Company A' ,'Company B' ])
```
