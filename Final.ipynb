แก้missing โดยการเติมค่าอื่นๆ

### Groupby
นำข้อมูลเข้าตารางgroupby และปรับข้อมูลด้วยค่าmean

```python
cv= df.groupby('Country').mean().reset_index()
```

```python
cv
```

#### เนื่องจากสนใจข้อมูลอันดับ2 ถึง อันดับ 10 จึงทำการคัดเลือกมาวิเคราะห์

```python
cvv = cv.iloc[1:10,:]
```

```python
cvv
```

### Plot Graph

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
cvv = cv.iloc[1:10,:]
```

```python
cvv
```

กราฟแสดงผลจำนวนผู้ติดเชื้อโควิดในแต่ละประเทศ

```python
plt.bar([1,2,3,4,5,6,7,8,9],cvv.iloc[:,2],tick_label = cvv.iloc[:,0],color = 'grey')
plt.title(u'กราฟเปรียบเทียบจำนวนผู้ติดเชื้อโควิดในแต่ละประเทศ',fontname='Tahoma',fontsize='13')
plt.xlabel('Country')
plt.ylabel('Confirmed')
```

กราฟแสดงการเปรียบเทียบผู้ได้รับการรักษาอาการติดเชื้อในแต่ละประเทศ

```python
plt.bar([1,2,3,4,5,6,7,8,9],cvv.iloc[:,4],tick_label = cvv.iloc[:,0],color = 'yellow')
plt.title(u'กราฟเปรียบเทียบจำนวนผู้ได้รับการรักษาอาการติดเชื้อในแต่ละประเทศ',fontname='Tahoma',fontsize='13')
plt.xlabel('Country')
plt.ylabel('Recovered')
```

กราฟแสดงการเปรียบเทียบผู้ตายจากเชื้อโควิดในแต่ละประเทศ

```python
plt.bar([1,2,3,4,5,6,7,8,9],cvv.iloc[:,3],tick_label = cvv.iloc[:,0],color = 'green')
plt.title(u'กราฟเปรียบเทียบจำนวนผู้ตายจากการติดเชื้อวิดในแต่ละประเทศ',fontname='Tahoma',fontsize='13')
plt.xlabel('Country')
plt.ylabel('Deaths')
```
