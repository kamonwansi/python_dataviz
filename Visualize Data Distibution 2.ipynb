ลบช่องว่างข้างหน้าชื่อออกเพื่อที่จะสามารถทำให้ไฟล์ 2 ไฟล์นั้นสามารถเชื่อมกันได้

```python
country_geo = 'thai.json'
```

```python
new_col =list()
for d in Data_to_plot['Region and province']:
    new_col.append(d.lstrip())
Data_to_plot['Region and province']=new_col
```

การกำหนดพิกัด ลองติจูดและละติจูดของประเทศไทย

```python
map = folium.Map(location=[10,100], zoom_start=5)
```

```python
map
```

ลงสีของข้อมูลที่มีค่าของจำนวนหนี้สินเฉลี่ยต่อครัวเรือนถ้าจังหวัดไหนที่มีจำนวนหนี้สินเฉลี่ยต่อครัวเรือนมีค่ามากจะเป็นสีม่วงเข้มส่วนจังหวัดไหนที่มีจำนวนหนี้สินเฉลี่ยต่อครัวเรือนน้อยจะเป็นสีเหลืองตามแถบรูปข้างบน

```python
map.choropleth(geo_data=country_geo, data=Data_to_plot,
              columns=['Region and province', 2549],
              key_on='feature.properties.name',
              fill_color='YlGnBu',fill_opacity=0.7, line_opacity=0.2,
              legend_name='Average Debt')
```
